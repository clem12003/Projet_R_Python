# Import the data

```python
import pandas as pd
import matplotlib.pyplot as plt

bazR = pd.read_csv("data/bazR_scores.csv")
```

# Calculation of z scores of each dimension with reference values of the French population (3 month)

```python
# Physical Functioning
bazR["PF_Z_3m"] = (bazR["PFtransformed_3m"] - 85.3) / 22.3
print(bazR["PF_Z_3m"].describe())

# Role Physical
bazR["RP_Z_3m"] = (bazR["RPtransformed_3m"] - 82.2) / 32.2
print(bazR["RP_Z_3m"].describe())

# Bodily Pain
bazR["BP_Z_3m"] = (bazR["BPtransformed_3m"] - 73.0) / 24.6
print(bazR["BP_Z_3m"].describe())

# General Health
bazR["GH_Z_3m"] = (bazR["GHtransformed_3m"] - 67.8) / 18.9
print(bazR["GH_Z_3m"].describe())

# Vitality
bazR["VT_Z_3m"] = (bazR["VTtransformed_3m"] - 57.4) / 18.0
print(bazR["VT_Z_3m"].describe())

# Social Functioning
bazR["SF_Z_3m"] = (bazR["SFtransformed_3m"] - 80.9) / 21.2
print(bazR["SF_Z_3m"].describe())

# Role Emotional
bazR["RE_Z_3m"] = (bazR["REtransformed_3m"] - 82.0) / 32.9
print(bazR["RE_Z_3m"].describe())

# Mental Health
bazR["MH_Z_3m"] = (bazR["MHtransformed_3m"] - 66.7) / 17.7
print(bazR["MH_Z_3m"].describe())
```

# Calculation of the Physical and Mental Component scores (3 month)

```python
# Physical Component Summary (PCS)
bazR["PhysicalComponentScale_z_3m"] = (
    (
        (bazR["PF_Z_3m"] * 0.42402) +
        (bazR["RP_Z_3m"] * 0.35119) +
        (bazR["BP_Z_3m"] * 0.31754) +
        (bazR["SF_Z_3m"] * -0.00753) +
        (bazR["MH_Z_3m"] * -0.22069) +
        (bazR["RE_Z_3m"] * -0.19206) +
        (bazR["VT_Z_3m"] * 0.02877) +
        (bazR["GH_Z_3m"] * 0.24954)
    ) * 10
) + 50

print(bazR["PhysicalComponentScale_z_3m"].describe())


# Mental Component Summary (MCS)
bazR["MentalComponentScale_z_3m"] = (
    (
        (bazR["PF_Z_3m"] * -0.22999) +
        (bazR["RP_Z_3m"] * -0.12329) +
        (bazR["BP_Z_3m"] * -0.09731) +
        (bazR["SF_Z_3m"] * 0.26876) +
        (bazR["MH_Z_3m"] * 0.48581) +
        (bazR["RE_Z_3m"] * 0.43407) +
        (bazR["VT_Z_3m"] * 0.23534) +
        (bazR["GH_Z_3m"] * -0.01571)
    ) * 10
) + 50

print(bazR["MentalComponentScale_z_3m"].describe())
```
